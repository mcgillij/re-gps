# Checking the format of the data from google export

```python
import json
import functools


@functools.lru_cache
def get_locations(file_to_open): 
    with open(file_to_open, 'r') as f:
        json_file = json.load(f)
        return [l for l in json_file["locations"]]

@functools.lru_cache
def get_json(file_to_open): 
    with open(file_to_open, 'r') as f:
        return json.load(f)
```

```python
file_to_open = 'sample-data/Records.json'
json_data = get_json(file_to_open)

locations = get_locations(file_to_open)
```
